```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# üìç DOƒûRU DOSYA YOLU (Desktop'tan itibaren)
csv_path = "Desktop/data_analyst_job_analysis/data/data_analyst_jobs.csv"

# üìÇ √áalƒ±≈üma dizinini yazdƒ±r
print("üìÇ √áalƒ±≈üma dizini:", os.getcwd())

# üìÑ Dosya var mƒ± kontrol√º
print("üìÑ Dosya var mƒ±?:", os.path.isfile(csv_path))

# ‚úÖ CSV dosyasƒ±nƒ± oku
df = pd.read_csv(csv_path)

# üîç ƒ∞lk 5 satƒ±rƒ± yazdƒ±r
print("\nüìå ƒ∞lk 5 Satƒ±r:")
print(df.head())

# üìê Veri seti boyutu
print("\nüìê Veri boyutu:", df.shape)

# üß± S√ºtun bilgileri ve veri tipleri
print("\nüß± S√ºtun bilgileri ve veri tipleri:")
print(df.info())

# ‚ö†Ô∏è Eksik deƒüer kontrol√º
print("\n‚ö†Ô∏è Eksik deƒüer kontrol√º:")
print(df.isnull().sum())

# ‚òÅÔ∏è A√ßƒ±klamalardan metin olu≈ütur
text = " ".join(df["description"].astype(str))

# ‚òÅÔ∏è WordCloud olu≈ütur
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

# üé® G√∂rselle≈ütir
plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title(" Job Description WordCloud")
plt.show()

# üíæ G√∂rseli kaydet
wordcloud.to_file("Desktop/data_analyst_job_analysis/visuals/job_description_wordcloud.png")
```
